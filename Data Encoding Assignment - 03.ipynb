# Ordinal Encoder With Loop

```python
df9.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
coll = df9.columns
coll
```

```python
for All in coll:
    if is_numeric_dtype(df9[All]):
        continue
        
    else:
        ord_value = OrdinalEncoder(categories=[df9[All].unique()])
        encod_value = ord_value.fit_transform(df9[[All]])
        encod_value = pd.DataFrame(encod_value,columns=[All])
        df9 = pd.concat([df9.drop(All,axis=1),encod_value],axis=1)
```
